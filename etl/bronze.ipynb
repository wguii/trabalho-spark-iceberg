```python
from pyspark.sql.functions import col

source_table = "sua_tabela"

df_landing = spark.table(f"landing.{source_table}")

df_bronze = df_landing.dropna(subset=["id"]) \
                      .withColumn("criado_em", col("criado_em").cast("timestamp"))

df_bronze.write.format("delta").mode("overwrite").saveAsTable(f"bronze.{source_table}_cleaned")
```
