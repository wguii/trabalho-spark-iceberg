```python
from pyspark.sql.functions import sum



source_table = "sua_tabela"


df_silver = spark.table(f"silver.{source_table}_curated")


df_gold = df_silver.groupBy("categoria").agg(
    sum("valor").alias("valor_total")
)


df_gold.write.format("delta").mode("overwrite").saveAsTable(f"gold.{source_table}_summary")

```
